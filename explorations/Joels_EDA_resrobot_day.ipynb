```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
from backend.resrobot_day import ResRobotDay
```

```python
# Skapar instans av ResRobotDay
rr_today = ResRobotDay()

station_id = 740000002

# Hämta avgångar och ankomster
df_dep = rr_today.departures_until_now(station_id)
df_arr = rr_today.arrivals_until_now(station_id)

print(f"Antal avgångar: {len(df_dep)}")
print(f"Antal ankomster: {len(df_arr)}")
```

```python
# Inspekterar DataFrame för Avgångar
print("\n=== DataFrame för Avgångar ===")
print(df_dep.head())
print("Kolumner:", df_dep.columns.tolist())
```

```python
# Inspekterar DataFrame för Ankomster
print("\n=== DataFrame för Ankomster ===")
print(df_arr.head())
print("Kolumner:", df_arr.columns.tolist())
```

```python
# Inspekterar unika avgångstider
if not df_dep.empty and "time" in df_dep.columns:
    print("\n=== Unika Avgångstider ===")
    print(df_dep["time"].unique())
else:
    print("\nIngen data för avgångar eller kolumn 'time' saknas i df_dep.")
```

```python

# Inspekterar unika ankomsttider
if not df_arr.empty and "time" in df_arr.columns:
    print("\n=== Unika Ankomstider ===")
    print(df_arr["time"].unique())
else:
    print("\nIngen data för ankomster eller kolumn 'time' saknas i df_arr.")
```

```python
# Hanterar avgångar
if not df_dep.empty and "time" in df_dep.columns:
    # Extraherar timme ur "time" (ex: "14:45:00" -> 14)
    df_dep["Hour"] = df_dep["time"].str.split(":").str[0].astype(int)
    
    # Räkna hur många avgångar per timme
    dep_count_by_hour = df_dep["Hour"].value_counts().sort_index()
    
    print("\n=== Distribution of Departures by Hour ===")
    print(dep_count_by_hour)
else:
    print("\nData saknas eller kolumn 'time' finns inte i df_dep.")
    dep_count_by_hour = pd.Series([0]*24, index=range(24))
```

```python
# Hantera ankomster
if not df_arr.empty and "time" in df_arr.columns:
    # Extraherar timme ur "time" (ex: "14:45:00" -> 14)
    df_arr["Hour"] = df_arr["time"].str.split(":").str[0].astype(int)
    
    # Räkna hur många ankomster per timme
    arr_count_by_hour = df_arr["Hour"].value_counts().sort_index()
    
    print("\n=== Distribution of Arrivals by Hour ===")
    print(arr_count_by_hour)
else:
    print("\nData saknas eller kolumn 'time' finns inte i df_arr.")
    arr_count_by_hour = pd.Series([0]*24, index=range(24))
```

```python
# Ser till att alla timmar 0-23 är representerade
hours = range(24)
dep_list = [dep_count_by_hour.get(hour, 0) for hour in hours]
arr_list = [arr_count_by_hour.get(hour, 0) for hour in hours]

# Skapa en DataFrame med båda
df_hours = pd.DataFrame({
    "Hour": hours,
    "Departures": dep_list,
    "Arrivals": arr_list
}).set_index("Hour")

print("\n=== df_hours ===")
print(df_hours)
```

```python
import seaborn as sns

# Omstrukturera DataFrame till långt format för Seaborn
df_long = df_hours.reset_index().melt(id_vars='Hour', value_vars=['Departures', 'Arrivals'],
                                      var_name='Type', value_name='Count')

# Visualisera båda i samma graf med Seaborn som grupperade staplar
sns.set(style="whitegrid")
plt.figure(figsize=(14,7))

sns.barplot(x='Hour', y='Count', hue='Type', data=df_long, palette={'Departures': 'skyblue', 'Arrivals': 'orange'})

plt.title("Avgångar & Ankomster per Timme (idag, 0..nu)", fontsize=16)
plt.xlabel("Timme (0-23)", fontsize=14)
plt.ylabel("Antal", fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Typ')
plt.tight_layout()
plt.show()


```
